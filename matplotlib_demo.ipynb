Histogram

```python
sugar=[23,34,24,12,46]
man=[4,32,13,23,45]

plt.hist(sugar,bins=[10,20,30,41,45,50], color='green', histtype='step', rwidth=0.7)
```

```python
plt.hist(sugar,bins=4, color='r', rwidth=0.7)
```

```python
plt.hist([sugar,man],color=['green','orange'], label=['sugar','man'] ,rwidth=0.7) #use '[]' array for the list to use two list in histogram
plt.legend()
plt.xlabel('sugar')
plt.ylabel('freq')
plt.title('Data')
```

```python
plt.hist([sugar,man],color=['green','orange'], orientation='horizontal', label=['sugar','man'] ,rwidth=0.7) #use '[]' array for the list to use two list in histogram
```

Pie Chart

```python
value=[100,400,600,410,240]
name=['rent','food','call','bill','other']
plt.pie(value,labels=name)

plt.axis("equal")
```

```python
plt.pie(value,labels=name, radius=1.5, autopct='%0.2f%%',shadow=True,) #autopct used to give percentage, specify the decimal value after 1st %, 
plt.show()
```

```python
plt.pie(value,labels=name, radius=1.5, autopct='%0.0f%%') #only percentage
plt.show()
```

```python
plt.pie(value,labels=name, radius=1.5, autopct='%0.2f%%',shadow=True,explode=[0,0,0,0.2,0]) #using explode at that position i.e. here 4 the 4th element will come out.
#here 1st elemet is 'rent' & number rotate anti clockwise
plt.show()
```

```python
plt.pie(value,labels=name, radius=1.5, autopct='%0.2f%%',shadow=True,startangle=45)#1st element ie rent get angle at 45
plt.show()
```

Save this charts to a file using savefig

```python
plt.pie(value,labels=name, radius=1.5, autopct='%0.2f%%',shadow=True,startangle=45)#1st element ie rent get angle at 45
plt.show()
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=2,transparent=True)
```
